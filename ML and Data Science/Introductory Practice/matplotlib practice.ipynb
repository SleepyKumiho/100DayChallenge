```python
# Practice on another dataset
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,30))
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(kind = "scatter", x = "age", y = "chol", c = "target")
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = "scatter", x = "age", y = "chol", c="target", ax=ax);

ax.set_xlim([45, 100])
```

```python
# Full OO method from pandas
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"], 
                     y = over_50["chol"], 
                     c=over_50["target"])
# Customize
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol level")

# Add a legend 
# query the scatter variable for the legend elements (c in this case), and provide a legend to match the possible values
ax.legend(*scatter.legend_elements(), title = "Target")

# Horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x = over_50["thalach"],
                     y = over_50["chol"],
                     c = over_50["target"])

ax.set(title = "Max Heart Rate and Cholesterol Levels",
       xlabel = "Max Heart Rate",
       ylabel = "Cholesterol Level")

ax.legend(*scatter.legend_elements(), title = "Target")
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10))
ax0.scatter(x = over_50["age"],
            y = over_50["chol"],
            c = over_50["target"])
ax0.set(title = "Age and Cholesterol Level",
        xlabel = "Age",
        ylabel = "Cholesterol Level")
ax0.legend(*scatter.legend_elements(), title = "Target")
ax0.axhline(over_50["chol"].mean(), linestyle = "--")

ax1.scatter(x = over_50["age"],
            y = over_50["thalach"],
            c = over_50["target"])
ax1.set(title = "Age and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title = "Target")
ax1.axhline(over_50["thalach"].mean(), linestyle = "--")

# Add a title to the full figure 
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

# Full OO method from pandas
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"], 
                     y = over_50["chol"], 
                     c=over_50["target"],
                    cmap = "winter") # changes the color scheme
# Customize
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol level")

# Add a legend 
# query the scatter variable for the legend elements (c in this case), and provide a legend to match the possible values
ax.legend(*scatter.legend_elements(), title = "Target")

# Horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
# Customizing x and y axis limits

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10),
                              sharex = True) # This tells the plot that the subplots share x values and removes some of the numerical instances. The label 
                                            # was removed manually. To be clear, this is just an available option, not a requirement of subplots.
scatter = ax0.scatter(x = over_50["age"],
            y = over_50["chol"],
            c = over_50["target"],
           cmap="winter")
ax0.set(title = "Age and Cholesterol Level",
        ylabel = "Cholesterol Level")
# Setting axis limits
ax0.set_xlim([50,80])

ax0.legend(*scatter.legend_elements(), title = "Target")
ax0.axhline(over_50["chol"].mean(), linestyle = "--")

scatter = ax1.scatter(x = over_50["age"],
            y = over_50["thalach"],
            c = over_50["target"],
           cmap="winter")
ax1.set(title = "Age and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")
# Changing axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

ax1.legend(*scatter.legend_elements(), title = "Target")
ax1.axhline(over_50["thalach"].mean(), linestyle = "--")

# Add a title to the full figure 
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
```
